# Creation of Dataframe

```python

# Split the content into individual candidate sections
candidates = content.strip().split("________________________________________")

# Initialize an empty list to hold each candidate's data
candidate_data = []

# Define a regular expression to capture candidate responses
response_pattern = re.compile(r'Response:\n"(.*?)"', re.DOTALL)

# Process each candidate's section
for candidate in candidates:
    candidate_info = {}
    
    # Extract candidate name
    candidate_info['name'] = re.search(r'Candidate \d+: (.*)', candidate).group(1)
    
    # Extract role
    candidate_info['Role Applied For'] = re.search(r'Role Applied For: (.*)', candidate).group(1)
    
    # Extract years of experience
    candidate_info['Years of Experience'] = re.search(r'Years of Experience: (\d+) years', candidate).group(1)
    
    # Find all the responses and combine them
    responses = response_pattern.findall(candidate)
    combined_responses = ' '.join(responses)
    
    candidate_info['Combined_Responses'] = combined_responses
    candidate_data.append(candidate_info)

# Create a DataFrame from the processed candidate data
df = pd.DataFrame(candidate_data)

# Display the DataFrame
print(df)
```

# Model Evaluation

```python
# Create a simple scoring system based on years of experience
df['experience_score'] = df['Years of Experience'].astype(float) / 5  # Normalize to 0-1 range

# Feature extraction (TF-IDF Vectorizer)
vectorizer = TfidfVectorizer(stop_words='english', max_features=100)
X = vectorizer.fit_transform(df['Combined_Responses'])

# Combine TF-IDF features with experience score
X_combined = np.hstack((X.toarray(), df['experience_score'].values.reshape(-1, 1)))

# Create a better suitability score by separating experience score from the target (less leakage)
np.random.seed(42)
y = np.random.rand(len(df)) * 0.5 + df['experience_score'] * 0.3  # Use less experience score, more random component

# Train the regression model using cross-validation with fewer folds
model = RandomForestRegressor(n_estimators=100, random_state=42)
scores = cross_val_score(model, X_combined, y, cv=2, scoring='r2')  # Using 2-fold cross-validation

# Report the results
print(f"Cross-validated R-squared scores: {scores}")
print(f"Mean R-squared: {np.mean(scores):.4f}")

# Fit the model on the full dataset for ranking purposes
model.fit(X_combined, y)

# Feature importance
feature_importance = model.feature_importances_
feature_names = vectorizer.get_feature_names_out().tolist() + ['experience_score']

# Plot feature importance
plt.figure(figsize=(12, 6))
sorted_idx = np.argsort(feature_importance)
sorted_idx = sorted_idx[-20:]  # Get top 20 features
pos = np.arange(sorted_idx.shape[0]) + .5
plt.barh(pos, feature_importance[sorted_idx], align='center')
plt.yticks(pos, [feature_names[i] for i in sorted_idx])
plt.xlabel('Feature Importance')
plt.title('Top 20 Most Important Features')
plt.tight_layout()
plt.savefig('feature_importance.png')
plt.close()

# Predict suitability scores for all candidates
df['predicted_score'] = model.predict(X_combined)

# Rank candidates
df['rank'] = df['predicted_score'].rank(ascending=False, method='min')
df_ranked = df.sort_values('predicted_score', ascending=False)

print("\nCandidate Rankings:")
print(df_ranked[['name', 'Role Applied For', 'Years of Experience', 'predicted_score', 'rank']])

print("Done")
```

### Suggestions to improve the model

1. Expand the Dataset:
* Add more candidates to the dataset to help the model generalize better. With only 4 samples, the model lacks sufficient data to learn patterns effectively. Aim for at least 50+ samples, if possible.
2. Improve Feature Engineering:
* Use advanced text representation techniques like Word2Vec or BERT embeddings instead of just TF-IDF. This will allow the model to capture richer semantic meaning from the candidates' responses.
* Add interaction features, such as combining years of experience with the complexity of answers (technical terms, length of response, etc.).
3. Optimize Model Selection:
* Instead of just using RandomForestRegressor, try more sophisticated models like Gradient Boosting (e.g., XGBoost or LightGBM), which often perform better on structured/tabular data.
* Use Logistic Regression or SVM for classification-based suitability scoring (if you decide to classify candidates as "suitable" or "not suitable").
4. Hyperparameter Tuning:
* Apply Grid Search or Random Search to find the best hyperparameters for your model. Tune parameters like the number of trees, learning rates, and maximum depth for better performance.
5. Cross-validation for Reliable Metrics:
* Implement k-fold cross-validation (e.g., 5-fold) to get more reliable performance metrics. This will ensure that your model is evaluated on multiple data splits and helps reduce variance in model performance.
